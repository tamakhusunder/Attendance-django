```python
# # Import the required libraries
# import random
import os
import cv2
import matplotlib.pyplot as plt
import albumentations as A
import glob    #to access image for any folder

# Load an image from the disk
Folder_to_access_photo="2016batch"
for filename in glob.glob(f'{Folder_to_access_photo}/*.png'):
    image = cv2.imread(filename)
    filename=filename[10:16]
    print(filename)
    print(image)

    # properties of albumentation
    transform0 = A.Compose([
        A.Flip(p=0.25),
        A.RandomBrightnessContrast(p=0.85),
        A.Rotate(limit=35, p=0.9),
        A.RandomRotate90(p=0.25),
        A.RGBShift(p=0.75),
    ])
    transform1 = A.Compose([A.ToSepia(p=1)])
    transform2 = A.Compose([A.ToGray(p=1)])
    transform3 = A.Compose([A.InvertImg(p=1)])
    # create a folder
    IMAGES_PATH = 'dataset2/dataset/'+filename
    os.makedirs(IMAGES_PATH, exist_ok=True)
    
    # Function for saving transformed figure in directory
    def saving_transformation(i,transform):
        augmented = transform(image=image)['image']
        file_name = f'{filename}_{i}.jpg'
        cv2.imwrite(f'{IMAGES_PATH}/{file_name}', augmented)
    saving_transformation(1,transform1)
    saving_transformation(2,transform2)
    saving_transformation(3,transform3)
    for i in (range(4,100+1)):
        saving_transformation(i,transform0)
```
